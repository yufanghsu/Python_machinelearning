```python
df = pd.read_csv('http://data.ntpc.gov.tw/od/data/api/5B69326E-55E3-47EC-8CE3-D4B9D06303F5?$format=csv')
colnames = df.loc[1].values
df = df[1:]
df.head()
dfrank = df.sort_values(by = ['Total'], ascending=False)
dfrank[:10].plot.bar(x='Scenic Spots', y=['Total'])            #新北十大熱門旅遊景點
dfrank[:10]
```

新北最熱門的景點在於瑞芳風景特定區和十分地區，也就是我們熟知的水金九地帶和平溪鐵路沿線。
