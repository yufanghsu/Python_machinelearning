```python
x=np.linspace(1,20,20)      #days
y=[173 ,38, 36, 31,29,22,11,8,6,10,4,17,2,6,4,4,4,3,1,2]   #event numbers
k=200
c=1
z=k/((c+x)**1.2)    #Omori's law
pit.plot(x,y,'.')
pit.plot(x,z,'r')
pit.ylabel("Numbers")
pit.xlabel("Time since Meinong mainshock occurred (days)")
pit.title("Relation between Time and Event Numbers")
pit.text(2.5, 60, 'N=k/((c+t)^1.2)')
```
