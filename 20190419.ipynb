# Loss Function & Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
import random
from sympy import *
```

## 產生函數 f(x)=x*cos(x)

```python
x=np.linspace(-10,10,200)
def func(x):
    return x*np.cos(x)
y=func(x)
plt.plot(x, y)         #函數繪圖
plt.xlabel('x')
plt.ylabel('f(x)')
w=random.uniform(-10, 10)                 #隨機取一點作為起始點
plt.plot(w, w*np.cos(w),'ro')         #起始點繪圖
```

## 定義gradient descent 函式

```python
## 建立gradient descent 函式：w=w-eta*dL/dw
eta=0.01     
x = Symbol('x')     # 定義名為x的變數
diff(x*cos(x), x)    # 微分後，斜率dL/dw=-x*sin(x) + cos(x)
```

```python
## 不斷迭代直到找到最小值
def GD(times, start_x, eta):
    first_x = w                   ## 起始點的 x,y 
    first_y = func(w)
    history_x =list()
    history_y =list()
    
    for i in range(0,times):
        history_x.append(first_x)
        history_y.append(first_y)
        first_x = first_x-eta*(-first_x*np.sin(first_x)+np.cos(first_x))        #w=w-eta*dL/dw
        first_y = func(first_x)        
    return (history_x, history_y)

results = GD(50,w,eta)     ##從初始值開始找，learn rate (eta)=0.01，迭代次數50次
```

## 繪製迭代結果

```python
xp = np.linspace(-10,10,200)
plt.plot(results[0], results[1],'ro', xp, func(xp))
plt.xlabel('x')
plt.ylabel('f(x)')
```

由於這個函數的local minimum太多，若初始點沒有落在適當的位置，很容易只找到local minimum
